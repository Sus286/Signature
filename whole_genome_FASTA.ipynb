```python
#GC CONTENT PER FILE FASTA INTERO

def calculate_gc_content(fasta_file_path):
    gc_count = 0
    total_bases = 0

    with open(fasta_file_path, 'r') as fasta_file:
        for line in fasta_file:
            if line.startswith('>'):
                continue  # Salta le righe degli header

            # Calcola il conteggio GC per la sequenza corrente
            gc_count += sum(1 for base in line.strip() if base in 'GCgc')
            total_bases += len(line.strip())

    # Calcola il contenuto GC per l'intera sequenza
    gc_content = (gc_count / total_bases) * 100

    return gc_content
```

```python
# Specifica il percorso del tuo file FASTA
fasta_file_path = '/content/drive/MyDrive/hg19.fa'

# Calcola il contenuto GC
gc_content = calculate_gc_content(fasta_file_path)

# Stampa il contenuto GC per l'intera sequenza
print(f'Contenuto GC per l\'intera sequenza: {gc_content:.2f}%')
```
