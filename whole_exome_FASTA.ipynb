```python
#estraggo esoni da hg19->lo faccio girare stanotte
!pip install biopython
from Bio import SeqIO

def extract_exon_sequences(fasta_file, tsv_file):
    exon_sequences = {}  # Dizionario per memorizzare le sequenze degli esoni

    # Leggi le informazioni sugli esoni dal file TSV
    with open(tsv_file, 'r') as tsv:
        next(tsv)  # Salta l'intestazione
        for line in tsv:
            parts = line.strip().split('\t')
            chrom = parts[0]
            exon_starts = list(map(int, parts[1].split(',')))
            exon_ends = list(map(int, parts[2].split(',')))

            # Leggi il file FASTA e estrai la sequenza del cromosoma
            with open(fasta_file, 'r') as fasta:
                for record in SeqIO.parse(fasta, 'fasta'):
                    if record.id == chrom:
                        # Estrai le sequenze degli esoni
                        exon_sequences.setdefault(chrom, [])
                        for start, end in zip(exon_starts, exon_ends):
                            exon_sequence = str(record.seq)[start:end]
                            exon_sequences[chrom].append(exon_sequence)

    return exon_sequences

# Specifica il percorso del tuo file FASTA e TSV
fasta_file_path = '/content/drive/MyDrive/hg19.fa'  # Sostituisci con il tuo percorso
tsv_file_path = '/content/drive/MyDrive/esoni.tsv'  # Sostituisci con il tuo percorso
```

```python
# Estrai le sequenze degli esoni
exon_sequences = extract_exon_sequences(fasta_file_path, tsv_file_path)
```

```python
# Stampa le sequenze degli esoni
for chrom, sequences in exon_sequences.items():
    print(f'Sequenze degli esoni per il cromosoma {chrom}:')
    for i, sequence in enumerate(sequences, start=1):
        print(f'Esone {i}: {sequence}')
```

```python
#devo calcolare il gc content
def calculate_gc_content(exon_sequences):
    all_sequences = ''.join(exon_sequences).upper()
    gc_count = sum(1 for base in all_sequences if base in ('G', 'C'))
    total_bases = len(all_sequences)
    gc_content = (gc_count / total_bases) * 100
    return gc_content
```

```python
gc_content = calculate_gc_content([sequence for sequences in exon_sequences.values() for sequence in sequences])
```
