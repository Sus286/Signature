```python
INPUT=Signatures_SuperFiltrato_WE_completo.T
```

```python
TARGET=Signatures_SuperFiltrato_completo.T
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import KFold
from torch.optim.lr_scheduler import ExponentialLR

def hellinger_distance(p, q):
    sqrt_p = torch.sqrt(p)
    sqrt_q = torch.sqrt(q)

    return  torch.sqrt(torch.sum((sqrt_p - sqrt_q)**2))/(2**0.5)


# Carica i tuoi dati di input e target
train_data = torch.tensor(INPUT.values)
train_labels = torch.tensor(TARGET.values)

# Configura il K-Fold
kf = KFold(n_splits=8, shuffle=True, random_state=42)

class ComplexNN(nn.Module):
    def __init__(self):
        super(ComplexNN, self).__init__()
        self.fc1 = nn.Linear(79, 256)
        self.fc2 = nn.Linear(256, 128)
        self.fc3 = nn.Linear(128, 64)
        self.fc4 = nn.Linear(64, 32)
        self.fc5 = nn.Linear(32, 32)
        self.fc6 = nn.Linear(32, 79)
        self.leaky_relu = nn.LeakyReLU(0.1)
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        x = self.leaky_relu(self.fc1(x))
        x = self.dropout(x)
        x = self.leaky_relu(self.fc2(x))
        x = self.leaky_relu(self.fc3(x))
        x = self.leaky_relu(self.fc4(x))
        x = self.leaky_relu(self.fc5(x))
        x = nn.functional.softmax(self.fc6(x), dim=1)
        return x

# Configura l'addestramento e il test della rete neurale
for fold, (train_idx, val_idx) in enumerate(kf.split(train_data)):
    print(f"\nFold {fold + 1}")
    batch_size = 16

    fold_train_data = train_data[train_idx]
    fold_train_labels = train_labels[train_idx]
    fold_val_data = train_data[val_idx]
    fold_val_labels = train_labels[val_idx]

    fold_train_dataset = TensorDataset(fold_train_data, fold_train_labels)
    fold_val_dataset = TensorDataset(fold_val_data, fold_val_labels)

    fold_train_loader = DataLoader(fold_train_dataset, batch_size=batch_size, shuffle=True)
    fold_val_loader = DataLoader(fold_val_dataset, batch_size=1, shuffle=False)

    model = ComplexNN()
    optimizer = optim.Adam(model.parameters(), lr=0.01)
    scheduler = ExponentialLR(optimizer, gamma=0.9)

    epochs = 100

    for epoch in range(epochs):
        scheduler.step()
        total_loss = 0.0
        model.train()
        for batch_data, batch_labels in fold_train_loader:
            optimizer.zero_grad()
            outputs = model(batch_data.float())
            hellinger_loss = hellinger_distance(outputs, batch_labels.float())
            loss = torch.mean(hellinger_loss)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()

        mean_loss = total_loss / len(fold_train_loader)
        print(f"Epoch {epoch + 1}/{epochs}, Training Hellinger Loss: {mean_loss}")

    model.eval()
    total_hellinger_loss = 0
    total_samples = 0

    with torch.no_grad():
        for val_data, val_labels in fold_val_loader:
            outputs = model(val_data.float())
            hellinger_loss_value = hellinger_distance(outputs, val_labels.float())
            total_hellinger_loss += hellinger_loss_value.item()
            total_samples += val_labels.size(0)

    average_hellinger_loss = total_hellinger_loss / total_samples
    print(f"Average Hellinger Loss on validation set: {average_hellinger_loss:.4f}")
```
