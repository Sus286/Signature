```python
import numpy as np

# Definisci una funzione per il confronto tra i blocchi di dati
def compare_blocks(block_BRCA, block_ESONI):
    merged_block = pd.merge(block_BRCA, block_ESONI, on='chrom')
    filtered_block = merged_block[
        (merged_block['pos_start'] >= merged_block['exonStarts']) &
        (merged_block['pos_end'] <= merged_block['exonEnds'])
    ]
    return filtered_block[['chrom', 'pos_start', 'pos_end']]

# Imposta la dimensione dei blocchi
block_size = 1000

# Calcola il numero di blocchi necessari per ciascun DataFrame
num_blocks_BRCA = int(np.ceil(len(BRCA_FR_SUPERFILTRATO_WG) / block_size))
num_blocks_ESONI = int(np.ceil(len(ESONI) / block_size))

# Dividi i DataFrame in blocchi e confrontali
for i in range(num_blocks_BRCA):
    start_idx_BRCA = i * block_size
    end_idx_BRCA = min((i + 1) * block_size, len(BRCA_FR_SUPERFILTRATO_WG))
    block_BRCA = BRCA_FR_SUPERFILTRATO_WG.iloc[start_idx_BRCA:end_idx_BRCA]

    for j in range(num_blocks_ESONI):
        start_idx_ESONI = j * block_size
        end_idx_ESONI = min((j + 1) * block_size, len(ESONI))
        block_ESONI = ESONI.iloc[start_idx_ESONI:end_idx_ESONI]

        result_block = compare_blocks(block_BRCA, block_ESONI)
        empty_df = empty_df.append(result_block, ignore_index=True)
```
