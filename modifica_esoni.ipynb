```python
import pandas as pd

# Specifica il percorso del file TSV
file_path = "/content/drive/MyDrive/esoni.tsv"

# Carica il file TSV in un DataFrame
esoni_copy = pd.read_csv(file_path, sep='\t')
```

```python
esoni_copy
```

```python
esoni_noduplicati=esoni_copy.drop_duplicates()
```

```python
ESONI=esoni_noduplicati.copy()
```

```python
esoni_noduplicati['exonEnds']=esoni_noduplicati['exonEnds']+1
esoni_noduplicati['exonStarts'] = esoni_noduplicati.apply(lambda row: row['exonStarts'] - 1 if row['exonStarts']!= 0 else row['exonStarts'], axis=1)
```

```python
# Specifica il percorso del file in cui desideri salvare il DataFrame in formato TSV
output_file_path = "/content/drive/MyDrive/esoni_noduplicati.tsv"

# Salva il DataFrame come file TSV
esoni_noduplicati.to_csv(output_file_path, sep='\t', index=False)
```
